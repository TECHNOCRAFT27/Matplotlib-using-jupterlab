# Matplotlib using jupterlab 

```python
import matplotlib.pyplot as plt
import numpy as np
```

Matplotlib graphs your data on Figures (e.g., windows, Jupyter widgets, etc.), each of which can contain one or more Axes, an area where points can be specified in terms of x-y coordinates (or theta-r in a polar plot, x-y-z in a 3D plot, etc.). The simplest way of creating a Figure with an Axes is using pyplot.subplots. We can then use Axes.plot to draw some data on the Axes, and show to display the figure:




A simple example

```python
x_data_value = [1,3,5,7,9]
y_data_value = [2,4,6,8,10]

plt.scatter(x_data_value,y_data_value)
plt.show()    # Show the figure.
```

when we want large random data we can use np

```python
# dataset
x_data_value = np.random.random(200)*100
y_data_value = np.random.random(200)*100

#plot point graph
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("point graph")

# take  (x,y)co-ordinare array, coluor, shape,size,transparent level
plt.scatter(x_data_value,y_data_value,c='green',marker='o',s=500,alpha=0.2)

# Show the figure
plt.show()   
```

```python
import matplotlib.pyplot as plt
import numpy as np

# dataset
language = ['python', 'html', 'css', 'js', 'c']
experience = np.random.random(5) * 100

plt.xlabel("Programming Languages")
plt.ylabel("Experience Level (%)")
plt.title("Experience by Language bar graph")

# plot bar graph
plt.bar(language, experience, color='w', align='center', width=0.1, edgecolor='black', lw=2)

# Show the figure
plt.show()
```

```python
# dataset
age=np.random.normal(20,1.5,1000)

plt.xlabel("age")
plt.ylabel("no of people")
plt.title("age histogram ")

# plot bar graph
plt.hist(age , bins=10)
# Show the figure
plt.show()
```

```python
# Generate sample age data: mean=20, std=1.5, n=1000
age = np.random.normal(20, 1.5, 1000)

# Plot histogram
plt.hist(age, bins=[age.min(), 15, 18, 21, 24, age.max()], edgecolor='black', color='skyblue',cumulative=True)

# Add title and axis labels
plt.title("Age Demographic with Histogram Graph")
plt.xlabel("Age Range")
plt.ylabel("Number of People")

# Show plot
plt.show()
```

```python
# Data
language = ['python', 'html', 'css', 'js', 'c']
vote = np.random.random(5) * 100

# Title
plt.title("Pie Chart")

# Plot Pie Chart
plt.pie(vote,
        labels=language,
        explode=[0, 0, 0, 0.1, 0],  
        autopct='%.2f%%',
        startangle=90,
        pctdistance=0.5,
       )
plt.legend()

# Show Plot
plt.show()
```

```python
from matplotlib import style
style.use("dark_background")


# Dataset
year = [2005 + x for x in range(21)]
income = np.random.random(21) * 100000

# Sort income
income.sort()

# Define Y-ticks
income_ticks = np.arange(int(income.min()), int(income.max()) + 1, 20000)

# Plot
plt.title("Income Report from 2005 to 2025")
plt.xlabel("Year")
plt.ylabel("Income in USD")
plt.yticks(income_ticks, [f'${x/1000:.0f}k' for x in income_ticks])  
plt.plot(year, income, marker='o', linestyle='-', color='blue')
plt.grid(True)

# Show
plt.show()
```

### custom theme

 Background & Layout Colors

| **Key**            | **Effect**                                                                 |
| ------------------ | -------------------------------------------------------------------------- |
| `axes.facecolor`   | Sets the **background color** of the plotting area (`#121212` = dark gray) |
| `figure.facecolor` | Sets the **background color of the entire figure** (outside the axes)      |
| `axes.edgecolor`   | Color of the **axes border lines**                                         |
| `legend.facecolor` | Background color of the **legend box**                                     |
| `legend.edgecolor` | Border color around the **legend box**                                     |

 Text and Label Colors

| **Key**            | **Effect**                                                               |
| ------------------ | ------------------------------------------------------------------------ |
| `axes.labelcolor`  | Color of **axis labels** (e.g., "X axis", "Y axis")                      |
| `xtick.color`      | Color of tick labels on the **X-axis**                                   |
| `ytick.color`      | Color of tick labels on the **Y-axis**                                   |
| `text.color`       | Default color for **text** (e.g., titles, annotations)                   |
| `font.size`        | Sets the **default font size** for all text (title, labels, ticks, etc.) |
| `axes.titleweight` | Makes axis titles **bold**                                               |

 Line and Grid Settings

| **Key**           | **Effect**                                                                 |
| ----------------- | -------------------------------------------------------------------------- |
| `lines.linewidth` | Width of plotted lines (e.g., line charts)                                 |
| `lines.color`     | Default color for lines (useful if not manually specified in `plt.plot()`) |
| `axes.grid`       | Turns **gridlines on** (`True`) by default                                 |
| `grid.color`      | Sets the **color of the grid lines**                                       |
| `grid.linestyle`  | Style of the grid lines (`--` = dashed)                                    |



```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# Define your custom style
custom_style = {
    'axes.facecolor': '#121212',
    'figure.facecolor': '#121212',
    'axes.edgecolor': '#ffffff',
    'axes.labelcolor': '#00ffcc',
    'xtick.color': '#cccccc',
    'ytick.color': '#cccccc',
    'grid.color': '#444444',
    'text.color': '#ffffff',
    'lines.linewidth': 2.5,
    'lines.color': '#00ffcc',
    'font.size': 12,
    'axes.titleweight': 'bold',
    'axes.grid': True,
    'grid.linestyle': '--',
    'legend.facecolor': '#222222',
    'legend.edgecolor': '#ffffff',
}

# Apply custom style
mpl.rcParams.update(custom_style)

# Sample plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.title("Custom Themed Plot")
plt.plot(x, y, label='sin(x)')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.show()
```

```python

# Dataset
shock_A = np.random.random(5) * 100
shock_B  = np.random.random(5) * 100
shock_C  = np.random.random(5) * 100

income = ['first','second','third']

# Plot
plt.title("stock of company")

plt.plot(shock_A, marker='o', linestyle='-', color='blue' , label='company_A')
plt.plot(shock_B, marker='o', linestyle='-', color='red' , label='company_B')
plt.plot(shock_C, marker='o', linestyle='-', color='green' , label='company_C')
plt.grid(True)

plt.legend(labels=income)
plt.legend(loc='lower center')

# Show
plt.show()

```

```python
x = np.arange(1, 180)  # Avoid 0 to prevent log(0)

fig, axs = plt.subplots(2, 2, figsize=(10, 6))

axs[0, 0].plot(x, np.sin(np.radians(x)))
axs[0, 0].set_title("Sine Graph")

axs[0, 1].plot(x, np.cos(np.radians(x)))
axs[0, 1].set_title("Cosine Graph")

axs[1, 1].plot(x, np.tan(np.radians(x)))
axs[1, 1].set_title("Tangent Graph")

axs[1, 0].plot(x, np.log(x))
axs[1, 0].set_title("Log Graph")

fig.suptitle('Trigonometry and Logarithm')
fig.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.savefig('trigonometry.png', dpi=300, transparent=True)
plt.show()
```

```python
# Create a new figure before adding 3D axes
fig = plt.figure()
axis3d = fig.add_subplot(111, projection='3d')

# Generate random data
x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

# 3D scatter plot
axis3d.scatter(x, y, z, c='b', marker='o' , s=500 , alpha=0.3)

# Set labels (optional)
axis3d.set_title('3D Scatter Plot')
axis3d.set_xlabel('X Axis')
axis3d.set_ylabel('Y Axis')
axis3d.set_zlabel('Z Axis')

plt.show()
```
